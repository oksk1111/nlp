# Numpy

```python
import numpy as np
```

## np.array()

```python
vec = np.array([1,2,3,4,5]) # dimension-1
print(vec) # value
print(type(vec)) # type
print(vec.ndim) # dim of vector
print(vec.shape) # size of vector
```

```python
mat = np.array([[1,2,3],[4,5,6]]) # dimension-2
print(mat)
print(type(mat))
print(mat.ndim)
print(mat.shape)
```

## ndarray 초기화

```python
zero_mat = np.zeros((2, 3)) # 모든값이 0인 배열
print(zero_mat)

one_mat = np.ones((2, 3)) # 모든값이 1인 배열
print(one_mat)

man_mat = np.full((2, 3), 5) # 모든값이 5인 배열
print(man_mat)

digonal_mat = np.eye(3) # 대각선으로 1인 배열
print(digonal_mat)

rand_mat = np.random.random((2, 2)) # 랜덤값으로 배열
print(rand_mat)
```

## arange, reshape

```python
# from 0 to 9
range_vec = np.arange(10) # like range(10)
print(range_vec)

# from 0 to 9, step +2
range_vec_n = np.arange(0, 10, 2) # like range(0, 10, 2)
print(range_vec_n)

# reshape
reshape_mat = np.array(np.arange(30).reshape(5, 6))
print(reshape_mat)
```

## Numpy 슬라이싱, 인덱싱

```python
mat = np.array([[1,2,3], [4,5,6]])
print(mat)

# row 1
slicing_mat = mat[0, :]
print(slicing_mat)

# row 2
slicing_mat = mat[1, :]
print(slicing_mat)

# row 1, column 0 데이터 인덱스
print(mat[1, 0])

# 여러 데이터 인덱싱
indexing_mat = mat[[1, 1], [0, 2]] # mat[1,0], mat[1,2] 데이터를 가져옴에 유의한다
print(indexing_mat)
```

## Numpy 연산

```python
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])

result = x + y # == np.add(x, y)
print(result)

result = y - x # == np.substract(y, x)
print(result)

result = x * 3 # == np.multiply(x, 3)
print(result)

result = x / 2 # == np.divide(x, 2)
print(result)
```

# Pandas

## Series

```python
import pandas as pd
```

```python
sr = pd.Series([10, 20, 30, 40], index=["A", "B", "C", "D"])
print(sr)
print("values: {}".format(sr.values))
print("index: {}".format(sr.index))
```

## Dataframe

```python
values = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
index = ["one", "two", "three"]
columns = ["A", "B", "C"]

df = pd.DataFrame(values, index=index, columns=columns)
print(df)
```

```python
# 리스트로 생성
data = [['10', 'AA', 1.1],
        ['20', 'BB', 2.2],
        ['30', 'CC', 3.3]]

df = pd.DataFrame(data)
print(df)
```

```python
# 딕셔너리로 생성
data = {'Name': ['AA', 'BB', 'CC'],
        'Age': ['10', '20', '30'],
        'Score': [1.1, 2.2, 3.3]}

df = pd.DataFrame(data)
print(df)
```

```python
print(df.head(2))
print(df.tail(2))
print(df['Name'])
```

# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
plt.title('test')
plt.plot([1, 2, 3, 4], [2, 4, 8, 6]) # x축 데이터, y축 데이터
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

```python
plt.title('student')
plt.plot([1, 2, 3, 4], [2, 4, 8, 6])
plt.plot([1.5, 2.5, 3.5, 4.5], [3, 5, 8, 10])
plt.xlabel('hours')
plt.ylabel('score')
plt.legend(['A student', 'B student']) # 범례
plt.show()
```
