# 원-핫 인코딩

단어 집합의 크기를 벡터의 차원으로 하고, 표현하고 싶은 단어의 인덱스에 1의 값을 부여하고, 다른 인덱스에는 0을 부여하는 단어의 벡터 표현 방식<br>

단어의 개수가 늘어날 수록 벨터를 저장하기 위한 공간이 늘어난다.<br>
단어의 유사도를 표현하지 못한다.<br><br>
> 다차원 공간에 벡터화하는 기법으로 단점 해결.
1. 카운트 기반의 벡터화 방법: LSA(잠재 의미 분석), HAL
2. 예측 기반 벡터화 방법: NNLM, RNNLM, Word2Vec, FastText
3. #1, #2 방법: GloVe

```python
from konlpy.tag import Okt

okt = Okt()
tokens = okt.morphs("나는 자연어 처리를 배운다")
print(tokens)
```

```python
word_to_index = {word: index for index, word in enumerate(tokens)}
print("단어 집합:", word_to_index)
```

```python
def one_hot_encoding(word, word_to_index):
    one_hot_vector = [0] * (len(word_to_index))
    index = word_to_index[word]
    one_hot_vector[index] = 1
    return one_hot_vector
```

```python
one_hot_encoding("자연어", word_to_index) # 형태소에서 특정 단어의 one-hot-encoding 위치
```

# 케라스를 이용한 원-핫 인코딩

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.utils import to_categorical

text = "나랑 점심 먹으러 갈래 점심 메뉴는 햄버거 갈래 갈래 햄버거 최고야"

tokenizer = Tokenizer()
tokenizer.fit_on_texts([text]) # encoding
print("단어 집합:", tokenizer.word_index)
```

```python
sub_text = "점심 먹으러 갈래 메뉴는 햄버거 최고야"
encoded = tokenizer.texts_to_sequences([sub_text])[0] # index sequences 로 변환
print(encoded)
```

```python
one_hot = to_categorical(encoded) # one-hot-encoding
print(one_hot)
```
