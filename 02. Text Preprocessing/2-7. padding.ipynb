최대 문장 길이 만큼 다른 문장들도 길이를 맞춰줄 것이며, 빈 공간을 패딩으로 채울것이다.

# Numpy로 패딩하기

```python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
preprocessed_sentences = [['barber', 'person'], ['barber', 'good', 'person'], ['barber', 'huge', 'person'], ['knew', 'secret'], ['secret', 'kept', 'huge', 'secret'], ['huge', 'secret'], ['barber', 'kept', 'word'], ['barber', 'kept', 'word'], ['barber', 'kept', 'secret'], ['keeping', 'keeping', 'huge', 'secret', 'driving', 'barber', 'crazy'], ['barber', 'went', 'huge', 'mountain']]
```

```python
tokenizer = Tokenizer()
# fit_on_texts() 안에 코퍼스를 입력으로 하면 빈도수를 기준으로 단어 집합을 생성
tokenizer.fit_on_texts(preprocessed_sentences)
encoded = tokenizer.texts_to_sequences(preprocessed_sentences)
print(encoded)
```

```python
# 최대 문장의 길이
max_len = max(len(item) for item in encoded)
print(max_len)
```

```python
for sentence in encoded: # 각 문장에 대해서
    while len(sentence) < max_len: # max_len 보다 작으면
        sentence.append(0)

padded_np = np.array(encoded)
print(padded_np)
```

# 케라스 전처리 도구로 패딩하기

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
encoded = tokenizer.texts_to_sequences(preprocessed_sentences)
print(encoded)
```

```python
padded = pad_sequences(encoded) # 패딩을 앞에 추가한다. 뒤에 추가하려면 padding='post' 옵션 필요
print(padded)

padded = pad_sequences(encoded, padding='post')
print(padded)
```

```python
# numpy, keras 결과가 동일함
print((padded == padded_np).all())
```

```python
# 길이에 제한을 두고 패딩하기
padded = pad_sequences(encoded, padding='post', maxlen=5) # 하진 이건 문장을 앞에서부터 잘랐다. 뒤에서 부터 자르고 싶다면 truncating='post'를 주어야 한다.
print(padded)

padded = pad_sequences(encoded, padding='post', maxlen=5, truncating='post')
print(padded)
```

```python
# padding 값을 다른 값으로 주고 싶을 때; 여기선 max_len + 1
last_value = len(tokenizer.word_index) + 1

padded = pad_sequences(encoded, padding='post', value=last_value)
print(padded)
```
